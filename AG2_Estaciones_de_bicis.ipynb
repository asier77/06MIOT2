### Importación de librerías

```python
# Importamos las librerías necesarias para la actividad
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dts
```

### Carga de los datos del archivo CSV en un DataFrame

```python
df_est = pd.read_csv('./estaciones_bici.csv', delimiter=';')
df_est.drop(df_est.columns[0], axis=1, inplace=True)
```

```python
df_est.describe()
```

```python
df_est.head()
```

### Obtención de número de estaciones con total de 30

```python
print(len(df_est[df_est['total'] == 30]))
df_est.query('total == 30').total.count()
df_est.loc[df_est['total'] == 30].total.count()
```

```python
# Mostramos el conjunto de estacion con total de 30
df_est_value_30 = df_est.loc[df_est['total'] == 30]
```

```python
df_est_value_30
```

### Número de estación con la media más alta de bicis disponibles

```python
df_est_value_30_media = df_est_value_30.groupby('estation').mean()
```

```python
df_est_value_30_media
```

```python
# lo ordenamos por el atributo 'available' descendentemente
df_est_value_30_media_orden = df_est_value_30_media.sort_values(by=['available'], ascending=False)
```

```python
df_est_value_30_media_orden
```

```python
# Obtenemos la estación con la media más alta de bicis disponibles
res_estacion = df_est_value_30_media_orden.head().index.values[0]
res_estacion
```

### Histograma de la estación de bicis disponibles

```python
# Mostraos los datos de la estación 143
df_estacion = df_est_value_30.loc[df_est_value_30['estation'] == res_estacion]
df_estacion
```

```python
# Histograma de las bicis disponibles de la estación 143
plt.hist(df_estacion['available'])
```

### Gráfica con la línea temporal de bicis disponibles

```python
figura = plt.plot(df_estacion.updated, df_estacion.available)
plt.xticks([0,19,39,59,79,99],rotation='vertical')
plt.show()
```
